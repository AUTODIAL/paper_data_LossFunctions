# Calculate the MAPE compared to the values used for generating data

```python
circuit_number = 0   # change this number from 0 to 5 to select each circuit at a time
all_circuits = ['R1-[C2,P3-R4]-L5',
 'R1-[P2,R3]',
 'R1-[P2,R3]-L4',
 'R1-[P2,R3]-P4',
 'R1-[P2,R3]-[P4,R5]',
 'R1-[P2,R3]-[P4,R5]-P6']

def param_error(loss_func_df, circuit_type, common_index_converged):
    loss_func_df = loss_func_df.iloc[common_index_converged]  # to make sure we have only the comon converged cases
    loss_function = loss_func_df
    circuit = circuit_type
    specific_circuit = loss_function[loss_function['circuit_string'] == circuit]
    counter = 0
    all_APE = []
    for row in specific_circuit.iterrows():
        row = row[1]
        counter += 1
        components = list(row.loc['param'].keys())
        estimated_values = np.array(list(row.loc['param'].values()))
        true_values = np.array(row.loc['true_param'])
        APE = np.abs(estimated_values - true_values) / np.abs(true_values) * 100
        all_APE.append(APE)

    df_APE = pd.DataFrame(np.array(all_APE), columns=[i for i in components])
    return df_APE

print(f'circuit: {all_circuits[circuit_number]}')
df_APE_X2 = param_error(metrics_X2, all_circuits[circuit_number], common_index_converged)
df_APE_PW = param_error(metrics_PW, all_circuits[circuit_number], common_index_converged)
df_APE_log_B = param_error(metrics_log_B, all_circuits[circuit_number], common_index_converged)
df_APE_log_BW = param_error(metrics_log_BW, all_circuits[circuit_number], common_index_converged)

all_MAPE = pd.concat([df_APE_X2.mean(), df_APE_PW.mean(), df_APE_log_B.mean(), df_APE_log_BW.mean()], keys=['X2', 'PW', 'log_B', 'log_BW'] , axis=1)
all_MAPE.plot(kind='bar', figsize=(10, 6), rot=45)
plt.title('Grouped Bar Plot')
plt.ylabel('MAPE (%)')
all_MAPE
```

```python

def time_plot(all_metrics, name, bin_width= 20, x_upper_lim = None, finename = None):
    titles = {'time': 'Time (s)'}

    # Prepare data in a long-form DataFrame for Seaborn
    data_list = []
    for metric in all_metrics:
        for value in all_metrics[metric][name]:
            data_list.append({'Metric': metric, f'{name}': value})

    df = pd.DataFrame(data_list)

    # Set up the figure and subplots
    fig, axes = plt.subplots(2, 1, figsize=(10, 12), gridspec_kw={'hspace': 0.3, 'wspace': 0.3})  # 2 rows, 1 column

    # Define a consistent color palette
    unique_metrics = df['Metric'].unique()
    palette = sns.color_palette('Set2', n_colors=len(unique_metrics))
    palette_dict = dict(zip(unique_metrics, palette))

    # Calculate bin edges based on the overall data range
    min_val = df[name].min()
    max_val = df[name].max()
    bin_edges = np.arange(min_val, max_val + bin_width, bin_width)

    # Plot histograms for each metric
    sns.histplot(data=df, x=name, hue='Metric',  kde=True, bins=bin_edges, ax=axes[0], alpha=0.6, palette=palette_dict)
    axes[0].set_xlabel(titles[name], fontsize=14, fontweight='bold')
    axes[0].set_ylabel('Frequency', fontsize=14, fontweight='bold')
    axes[0].set_title(f'{titles[name]} Histograms for All loss functions')
    axes[0].tick_params(axis='x', labelsize=12)  
    axes[0].tick_params(axis='y', labelsize=12)
    if x_upper_lim is not None:
        axes[0].set_xlim(0, x_upper_lim)

    # Plot box plots for each metric
    sns.boxplot(data=df, x=name, y='Metric', orient='h', ax=axes[1], palette=palette_dict, showmeans=True, meanprops={"marker": "o", "markerfacecolor": "red", "markeredgecolor": "black", "markersize": 8})
    axes[1].set_xlabel(titles[name], fontsize=14, fontweight='bold')
    axes[1].set_title(f'Time Box Plots for All loss functions')
    axes[1].tick_params(axis='x', labelsize=12)  
    axes[1].tick_params(axis='y', labelsize=12)
    if x_upper_lim is not None:
        axes[1].set_xlim(0, x_upper_lim)

    axes[1].set_ylabel('Loss Function', fontsize=14, fontweight='bold')    
    # Adjust layout and show the plot
    plt.tight_layout()
    plt.show()

    if finename is not None:
        fig.savefig(f'{finename}_{name}.jpg',  dpi=300, bbox_inches='tight')
```

```python

def histogram_and_box_plot(all_metrics, fit_quality, bin_width, x_lower_lim = None, x_upper_lim = None, finename = None):
    # Prepare data in a long-form DataFrame for Seaborn
    titles = {'r2_score': r'$\mathbf{R^2}$ Score', 'chi_square': r'chi_square', 'r2_mag': r'$\mathbf{R^2}$ Magnitude', 'r2_phase': r'$\mathbf{R^2}$ Phase', 'X2_internal': r'$\mathbf{Chi-squared}$'}
    data_list = []
    for metric in all_metrics.keys():
        for value in all_metrics[metric][fit_quality]:
            data_list.append({'Metric': metric, f'{fit_quality}': value})

    df = pd.DataFrame(data_list)


    # Set up the figure and subplots
    fig, axes = plt.subplots(2, 1, figsize=(10, 12), gridspec_kw={'hspace': 0.3, 'wspace': 0.3})  # 2 rows, 1 column

    # Define a consistent color palette
    unique_metrics = df['Metric'].unique()
    palette = sns.color_palette('Set2', n_colors=len(unique_metrics))
    palette_dict = dict(zip(unique_metrics, palette))

    # Calculate bin edges based on the overall data range
    min_val = df[fit_quality].min()
    max_val = df[fit_quality].max()
    bin_edges = np.arange(min_val, max_val + bin_width, bin_width)

    # Plot histograms for each metric
    sns.histplot(data=df, x=fit_quality, hue='Metric',  kde=True, bins=bin_edges, ax=axes[0], alpha=0.6, palette=palette_dict)
    # sns.histplot(data=df, x=fit_quality, hue='Metric', element='step',  kde=True, bins=bin_edges, ax=axes[0], alpha=0.3)
    axes[0].set_xlabel(titles[fit_quality], fontsize=14, fontweight='bold')
    axes[0].set_ylabel('Frequency', fontsize=14, fontweight='bold')
    axes[0].set_title(f'{titles[fit_quality]} Histograms for All loss functions')
    axes[0].tick_params(axis='x', labelsize=12)  
    axes[0].tick_params(axis='y', labelsize=12)
    if x_upper_lim is not None:
        axes[0].set_xlim(0, x_upper_lim)
    if x_lower_lim is not None:
        axes[0].set_xlim(x_lower_lim, 1)  

    # Plot box plots for each metric
    sns.boxplot(data=df, x=fit_quality, y='Metric', orient='h', ax=axes[1], palette=palette_dict, showmeans=True, meanprops={"marker": "o", "markerfacecolor": "red", "markeredgecolor": "black", "markersize": 8})
    axes[1].set_xlabel(titles[fit_quality], fontsize=14, fontweight='bold')
    axes[1].set_title(f'{titles[fit_quality]} Box Plots for All loss functions')
    axes[1].tick_params(axis='x', labelsize=12)  
    axes[1].tick_params(axis='y', labelsize=12)
    if x_upper_lim is not None:
        axes[1].set_xlim(0, x_upper_lim)
    if x_lower_lim is not None:
        axes[1].set_xlim(x_lower_lim, 1)    
    axes[1].set_ylabel('Loss Function', fontsize=14, fontweight='bold')    
    # Adjust layout and show the plot
    # plt.tight_layout()
    plt.show()

    if finename is not None:
        fig.savefig(f'{finename}_{fit_quality}.jpg',  dpi=300, bbox_inches='tight')


```
